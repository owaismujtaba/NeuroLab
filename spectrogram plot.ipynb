```python
from matplotlib import pyplot as plt
```

```python
import numpy as np

actual = np.load('Predictions/01/NeuralNetwork_R_1/Actual_Fold_0.npy')
nn = np.load('Predictions/01/NeuralNetwork_R_1/Predictions_Fold_0.npy')
lr = np.load('Predictions/01/LR/Predictions_Fold_0.npy')
actual.shape, nn.shape, lr.shape
```

```python
# Plot the actual spectrogram
start = 0
end = 500

actuals = actual[start:end]
nns = nn[start:end]
lrs = lr[start:end]

fig, axes = plt.subplots(1, 3, figsize=(15, 6), sharex=True, sharey=True)

axes[0].imshow(actuals.T, aspect='auto', origin='lower', cmap='viridis', interpolation='none')
axes[0].set_title('Actual', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Windows', fontsize=14, fontweight='bold')
axes[0].set_ylabel('Spectral Bins', fontsize=14, fontweight='bold')

# Plot predicted spectrogram
im = axes[1].imshow(nns.T, aspect='auto', origin='lower', cmap='viridis', interpolation='none')
axes[1].set_title('Neural Network', fontsize=14, fontweight='bold')

im = axes[2].imshow(lrs.T, aspect='auto', origin='lower', cmap='viridis', interpolation='none')
axes[2].set_title('Linear Regression', fontsize=14, fontweight='bold')

axes[0].tick_params(axis='both', labelsize=12)
axes[1].tick_params(axis='both', labelsize=12)
axes[2].tick_params(axis='both', labelsize=12)
#plt.xticks(rotation=45, fontsize=12)  # Increase font size to 14 (or another preferred size)
#plt.yticks(fontsize=12)
# Add colorbar to the second subplot
fig.colorbar(im, ax=axes[2], label='Spectral Intensity')

plt.tight_layout()
plt.savefig('Images/spectrograms_final.png', dpi=800)
```
