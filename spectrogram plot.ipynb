```python
from matplotlib import pyplot as plt
```

```python
import numpy as np

actual = np.load('/home/owaismujtaba/projects/NeuroLab/Predictions/22/Actual_Fold_0.npy')
predicted = np.load('/home/owaismujtaba/projects/NeuroLab/Predictions/22/Predictions_Fold_0.npy')
actual.shape, predicted.shape
```

```python
# Plot the actual spectrogram
start = 0
end = 500
for i in range(10):
    

    actuals = actual[start:end]
    predicteds = predicted[start:end]
    plt.figure(figsize=(10, 6))
    plt.subplot(1, 2, 1)
    plt.imshow(actuals.T, aspect='auto', origin='lower', cmap='viridis', interpolation='none')
    plt.title('Actual Spectrogram')
    plt.xlabel('windows')
    plt.ylabel('spectral bins')

    # Plot the predicted spectrogram
    plt.subplot(1, 2, 2)
    plt.imshow(predicteds.T, aspect='auto', origin='lower', cmap='viridis', interpolation='none')
    plt.colorbar(label='Spectral intensity')
    plt.title('Predicted Spectrogram')


    # Show the plots
    plt.tight_layout()
    plt.show()
    start = end
    end = end + 500
'''plt.savefig('Images/spectrograms.png', dpi=600)'''
```
